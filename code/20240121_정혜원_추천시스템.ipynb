# 추천 시스템
- 개인 고객 리뷰 데이터에 따라 top 10 category 추천
- 로그인 안 된 사람에 대해 리뷰 개수와 평균 별점에 대해 총 top 10 category 추천
- top 10 추천 카테고리 각각에 대해 top 2 상품으로 총 20개의 추천 상품 추천

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
```

```python
def category_recommendation(data):
    products = pd.read_csv('../json_csv_files/total.csv')
    products = products[['카테고리1', '카테고리2', '제품명', '리뷰 개수', '상품 별점']]
    products

    products_sample = products
    for idx, row in products_sample.iterrows():
        products_sample.loc[idx, '제품명'] = row['제품명'].split(',')[0]
    
        products = products_sample.drop_duplicates(subset='제품명', keep='first')
    
        second = products['카테고리2'].unique()
    second = [i.replace(' ', '') for i in second]

    for idx, row in data.iterrows():
        if (row['카테고리2'] not in second):
            data.drop([idx], axis=0, inplace=True)
    data = data.reset_index()

    data = data[['별점', '카테고리1', '카테고리2']]
    
    data_count = data[['별점', '카테고리1', '카테고리2']]
    data_count.rename(columns = {'별점': '리뷰 개수'}, inplace=True)
    
    rating = data.groupby(['카테고리1', '카테고리2']).mean().sort_values(by=['별점'], ascending=False)
    
    counts = data_count.groupby(['카테고리1', '카테고리2']).count().sort_values(by=['리뷰 개수'], ascending=False)
    
    rating_count = pd.concat([rating, counts], axis=1).sort_values(by=['리뷰 개수'], ascending=False)
    
    # 리뷰 개수: v
    # 평균 평점: r
    # 전체 리뷰에 대한 평균 별점: c
    # 가중 평점: v*r + v*c
    average_rating = rating_count['별점'].mean()
    rating_count['가중 평점'] = rating_count['리뷰 개수'] * rating_count['별점'] + rating_count['별점'] * average_rating
    rating_count = rating_count.sort_values(by=['가중 평점'], ascending=False)
    rating_count = rating_count.reset_index()
    
    top10_category = rating_count.loc[:9]

    return top10_category
```

```python
def products_recommendation(data):
    products = pd.read_csv('../json_csv_files/total.csv')
    products = products[['카테고리1', '카테고리2', '제품명', '리뷰 개수', '상품 별점']]
    products

    products_sample = products
    for idx, row in products_sample.iterrows():
        products_sample.loc[idx, '제품명'] = row['제품명'].split(',')[0]
    
        products = products_sample.drop_duplicates(subset='제품명', keep='first')
    
        second = products['카테고리2'].unique()
    second = [i.replace(' ', '') for i in second]

    for idx, row in data.iterrows():
        if (row['카테고리2'] not in second):
            data.drop([idx], axis=0, inplace=True)
    data = data.reset_index()

    data = data[['별점', '카테고리1', '카테고리2']]
    
    data_count = data[['별점', '카테고리1', '카테고리2']]
    data_count.rename(columns = {'별점': '리뷰 개수'}, inplace=True)
    
    rating = data.groupby(['카테고리1', '카테고리2']).mean().sort_values(by=['별점'], ascending=False)
    
    counts = data_count.groupby(['카테고리1', '카테고리2']).count().sort_values(by=['리뷰 개수'], ascending=False)
    
    rating_count = pd.concat([rating, counts], axis=1).sort_values(by=['리뷰 개수'], ascending=False)
    
    # 리뷰 개수: v
    # 평균 평점: r
    # 전체 리뷰에 대한 평균 별점: c
    # 가중 평점: v*r + v*c
    average_rating = rating_count['별점'].mean()
    rating_count['가중 평점'] = rating_count['리뷰 개수'] * rating_count['별점'] + rating_count['별점'] * average_rating
    rating_count = rating_count.sort_values(by=['가중 평점'], ascending=False)
    rating_count = rating_count.reset_index()
    
    top10_category = rating_count.loc[:9]
    top20_products = pd.DataFrame(columns = ['카테고리1', '카테고리2', '제품명', '리뷰 개수', '상품 별점'])
    for i in top10:
        category = pd.DataFrame(columns = ['카테고리1', '카테고리2', '제품명', '리뷰 개수', '상품 별점'])
        for idx,row in products.iterrows():
            if (row['카테고리1'].replace(' ', '') == i[0].replace(' ', '')) & (row['카테고리2'].replace(' ', '') == i[1].replace(' ', '')):
                row = pd.DataFrame([row], columns=products.columns)
                category = pd.concat([category, row])
        # 리뷰 개수: v
        # 상품 평점: r
        # 전체 리뷰에 대한 평균 별점: c
        # 가중 평점: (v*r)/c
        average_rating = category['상품 별점'].mean()
        category['가중 평점'] = (category['리뷰 개수'] * category['상품 별점']) / average_rating
        category = category.sort_values(by=['가중 평점'], ascending=False)
        category = category.drop_duplicates(subset='리뷰 개수', keep='first')
        category = category.reset_index()
        category = category.loc[:2]
        top20_products = pd.concat([top20_products, category])
                
    top20_products = top20_products.reset_index()

    return top20_products
```
