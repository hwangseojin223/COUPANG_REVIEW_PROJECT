```python
import pandas as pd
cat_review = pd.read_csv('../hdfs_data/total_data_ver2.csv', encoding='utf-8')
```

```python
cat_reviews = cat_review[cat_review.columns[:14]]
cat_reviews.columns
```

```python
cat_review.rename(columns={'카테고리1': 'category1', '카테고리2': 'category2', '카테고리 url': 'category_url', '제품명': 'product_name', '제품 url': 'product_url', '리뷰 개수': 'review_count', '상품 별점': 'product_rating', '리뷰': 'review'}, inplace = True)
```

```python
reviews = cat_review[['category1', 'category2', 'product_name', 'review_count', 'review', 'product_url']]
category1 = []
category2 = []
product = []
rating_cnt = []
name = []
member_id = []
rating = []
title_content = []
help = []
date = []
url = []

for idx, star in reviews.iterrows():
    ori_star = star
    star = star['review']
    try:
        star = eval(star)
        for r in star:
            for review in r:
                name.append(review[0])
                member_id.append(review[1])
                if review[2] == '5':
                    rating.append(3)
                elif review[2] == '4':
                    rating.append(2)
                else:
                    rating.append(1)
                title_content.append(str(review[3]) + ' ' + str(review[4]))
                help.append(review[5])
                date.append(review[6])
                category1.append(ori_star['category1'])
                category2.append(ori_star['category2'])
                product.append(ori_star['product_name'])
                rating_cnt.append(ori_star['review_count'])
                url.append(ori_star['product_url'])

        
    except:
        continue
df = pd.DataFrame(
    {'category1': category1,
     'category2': category2,
     'product_name': product,
     'product_url': url,
     'rating_cnt':rating_cnt,
     'reviewer_name': name,
     'userid': member_id,
     'ratings': rating,
     'review_contents': title_content,
     'help': help,
     'date': date},
)
df.to_csv('../hdfs_data/reviews_total.csv', index=False, encoding="utf-8-sig")
```
