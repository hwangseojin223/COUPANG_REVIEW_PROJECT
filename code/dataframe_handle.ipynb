```python
import pandas as pd

review_data_10 = pd.read_csv('../json_csv_files/customers10_data.csv', encoding='utf-8')
reviews_total = pd.read_csv('../hdfs_data/reviews_total.csv', encoding='utf-8')
review_data_10.columns
```

```python
reviews_total.columns
```

```python
review_data_10['product_url'] = 'None'
review_data_10['rating_cnt'] = 'None'
for idx, review in review_data_10.iterrows():
    if review['ratings'] == 5:
        review_data_10.loc[idx, 'ratings'] = 3
    elif review['ratings'] == 4:
        review_data_10.loc[idx, 'ratings'] = 2
    else:
        review_data_10.loc[idx, 'ratings'] = 1
review_data_10
```

```python
review_data_10 = review_data_10.fillna('None')
```

```python
total = pd.concat([review_data_10, reviews_total])
```

```python
total.to_csv('./reviews_total.csv', sep='+', line_terminator='@@', index=False)
```

```python
total = total[['category1', 'category2', 'product_name', 'product_url', 'rating_cnt',
       'reviewer_name', 'userid', 'ratings', 'help',
       'date', 'review_contents']]
```
